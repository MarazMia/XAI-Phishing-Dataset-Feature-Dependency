```python
import shap
import xgboost
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.metrics import classification_report
from sklearn.metrics import confusion_matrix
from sklearn.ensemble import RandomForestClassifier
```

```python
_80_20_traindf = pd.read_csv('../Dataset/Training/Dataset2/80_20_train.csv')
_80_20_testdf = pd.read_csv('../Dataset/Testing/Dataset2/80_20_test.csv')

print('-------------------------Using XGBoost---------------------------------\n')

_80_20_xgb_classifier = xgboost.XGBClassifier(max_depth=5, objective='binary:logistic', n_estimators=100)
_80_20_xgb_model = _80_20_xgb_classifier.fit(_80_20_traindf.drop(columns='status',axis=1), _80_20_traindf['status'])
_80_20_xgb_predicted_values = _80_20_xgb_model.predict(_80_20_testdf.drop(columns='status',axis=1))
print(classification_report(_80_20_testdf['status'],_80_20_xgb_predicted_values,target_names=['legitimate','phishing']))

print('-------------------------Using RandomForrest----------------------------\n')
_80_20_rf_classifier = RandomForestClassifier(max_depth=5, random_state=42)
_80_20_rf_model = _80_20_rf_classifier.fit(_80_20_traindf.drop(columns='status',axis=1), _80_20_traindf['status'])
_80_20_rf_predicted_values = _80_20_rf_model.predict(_80_20_testdf.drop(columns='status',axis=1))
print(classification_report(_80_20_testdf['status'],_80_20_rf_predicted_values,target_names=['legitimate','phishing']))
```
